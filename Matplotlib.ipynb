```
import numpy as np
t = np.arange(0.,5.0, 2.0)
print(t)
```

```
plt.plot(t, t**2, 'b--', label="blue")
plt.plot(t, t**2.2, 'rs', label="red")
plt.plot(t, t**2.5, 'g^', label='green')
plt.legend()
plt.show()
```

```
#set property = setp
x1=[1,2,3,4]
y1=[1,4,6,16]
x2=[1,2,3,4]
y2=[2,4,6,8]

lines= plt.plot(x1,y1, x2,y2)
print(lines)
plt.setp(lines[0], color='g', linewidth=2.0, label='Line 0')
plt.setp(lines[1], color='r', linewidth=3.0, label='Line 1') #MATLAB style

#plt.setp(lines[1], 'color', 'r', linewidhth=3.0, label='1') #MATLAB style does not work
plt.legend()
plt.show()
```
